```python
for col in df.columns:
    df.rename(columns={col:col.strip()},inplace=True)
    print(col)
    print(df[col].unique())
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(inplace=True)
```

```python
df.iloc[122]
```

```python
df = pd.concat([df.iloc[0:122], df.iloc[123:]], axis=0).reset_index(drop=True)
```

```python
df.reset_index()
```

```python
import numpy as np
```

```python
df['Region']=np.array(122*[0]+121*[1])
```

```python
df.shape
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
```

```python
for col in df.columns:
    if col.strip() == 'Classes':
        continue

    try:
        df[col] = df[col].astype(float)
        if (df[col] % 1 == 0).all():
            df[col] = df[col].astype(int)
            print(f"Column '{col}' converted to int.")
        else:
            print(f"Column '{col}' remains float.")
    except Exception as e:
        print(f"Could not convert column '{col}' to float. Exception: {e}")
```

```python
df.to_csv('cleaned_data.csv',index=False)
```
